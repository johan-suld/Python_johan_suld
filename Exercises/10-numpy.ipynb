## 1. Dice simulations (*)

Simulations of one die,  

&nbsp; a) Calculate theoretical mean of a dice (six-sided) (*)

&nbsp; b) Now do simulations of

- 10
- 100
- 1000
- 10000
- 100000
- 1000000
- 10000000

number of dices and plot their means. (*)

```python
import numpy as np
import matplotlib.pyplot as plt

print('a) 3.5')

simulation_sizes = [10, 100, 1000, 10000, 100000, 1000000, 10000000]

def dice_simulation(nr_of_sims):
    
    dice_rolls = []
    
    for i in range(nr_of_sims):
        dice_rolls.append(np.random.randint(1, 7))

    return dice_rolls

results = []

for size in simulation_sizes:
    results.append(np.mean(dice_simulation(size)))

plt.title('b) Mean of a six sided die')
plt.xticks([0, 1, 2, 3, 4, 5, 6], simulation_sizes)
plt.xlabel('Number of rolls')
plt.ylabel('Sample mean')
plt.plot(results, linestyle='dashed', marker='o')
```

## 2. Several dices (**)

&nbsp; a) Make a sample space for the sum of two dices, i.e. all possible outcomes of these dices. (*)

&nbsp; b) Count the number of each values in the sample space (*)

&nbsp; c) Now calculate the frequency $f_i$ for each unique value. Also check that the sum of frequencies sums to 1, i.e. $\sum_i f_i = 1$. (*)

&nbsp; d) Make a bar chart of the sum of two dices, with frequencies in y and unique values in x. This is a probability distribution of the sum of your two dices. (*)

&nbsp; e) Now make barplots in subplots of sum of dices for 1 dice, 2 dices, 3 dices and 4 dices. What is the pattern that you see? (**)

```python
import numpy as np
import matplotlib.pyplot as plt

def sum_dices(dices):

    sample_space = np.add.reduce(np.meshgrid(*[[1, 2, 3, 4, 5, 6]] * dices))

    unique, counts = np.unique(sample_space, return_counts=True)

    return unique, counts, sample_space

print(f'a)\n{sum_dices(2)[2]}')

print(f'\nb)\nUnique values:\n{sum_dices(2)[0]}\nCounts:\n{sum_dices(2)[1]}')

print(f'\nc)\n{sum_dices(2)[1] / sum(sum_dices(2)[1])}')

plt.bar(sum_dices(2)[0], sum_dices(2)[1] / sum(sum_dices(2)[1]))
plt.title('Probability distribution of sum of 2 dices')
plt.show

fig, axs = plt.subplots(2, 2)
axs[0, 0].bar(sum_dices(1)[0], sum_dices(1)[1] / sum(sum_dices(1)[1]))
axs[0, 1].bar(sum_dices(2)[0], sum_dices(2)[1] / sum(sum_dices(2)[1]))
axs[1, 0].bar(sum_dices(3)[0], sum_dices(3)[1] / sum(sum_dices(3)[1]))
axs[1, 1].bar(sum_dices(4)[0], sum_dices(4)[1] / sum(sum_dices(4)[1]))
axs[0, 0].title.set_text('Prob dist of sum of 1 die')
axs[0, 1].title.set_text('Prob dist of sum of 2 dices')
axs[1, 0].title.set_text('Prob dist of sum of 3 dices')
axs[1, 1].title.set_text('Prob dist of sum of 4 dices')
fig.tight_layout()
```
